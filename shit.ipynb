```python
import torch
import torch.nn as nn
from torch.nn import functional as F
import tiktoken
import urllib.request
from transformers import AutoTokenizer
import tiktoken
```

```python
text = "Реве та стогне Дніпр широкий"

print("========================================")
print("1. LLAMA-3 TOKENIZER (128k vocab)")
print("========================================")
llama_tokenizer = AutoTokenizer.from_pretrained("NousResearch/Meta-Llama-3-8B")
llama_tokens = llama_tokenizer.encode(text, add_special_tokens=False)

print(f"{'TOKEN ID':<10} | {'TEXT CHUNK'}")
print("-" * 30)
for tok_id in llama_tokens:
    chunk = llama_tokenizer.decode([tok_id])
    print(f"{tok_id:<10} | {repr(chunk)}")

print("\n========================================")
print("2. OPENAI o200k_base TOKENIZER (200k vocab)")
print("========================================")
# Load OpenAI's tokenizer
openai_tokenizer = tiktoken.get_encoding("o200k_base")
openai_tokens = openai_tokenizer.encode(text)

print(f"{'TOKEN ID':<10} | {'TEXT CHUNK'}")
print("-" * 30)
for tok_id in openai_tokens:
    # tiktoken requires decoding from bytes to handle partial UTF-8 safely
    raw_bytes = openai_tokenizer.decode_single_token_bytes(tok_id)
    chunk = raw_bytes.decode('utf-8', errors='replace')
    print(f"{tok_id:<10} | {repr(chunk)}")
```
